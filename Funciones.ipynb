Hay dos tipos de funciones en Python.
- Funciones integradas que se proporcionan como parte de Python - print (), input (), type (), float (), int () ...
- Funciones que definimos y luego usamos

Tratamos los nombres de las **funciones incorporadas** como palabras **reservadas** (es decir, las evitamos como nombres de variables)


En Python, una **función es un código reutilizable** que toma argumentos como entrada, realiza algunos cálculos y luego devuelve un resultado o resultados

- Definimos una función usando la palabra reservada **def**
- Llamamos / invocamos la función usando el nombre de la función, paréntesis y argumentos en una expresión

### Argumentos
- Un argumento es un valor que pasamos a la función como su entrada cuando llamamos a la función
- Usamos argumentos para poder dirigir la función para que haga diferentes tipos de trabajo cuando la llamamos en diferentes momentos.
- Ponemos los argumentos entre paréntesis después del nombre de la función




### Parámetros

- Un parámetro es una variable que usamos en la definición de función. 
- Es un "identificador" que permite que el código de la función acceda a los argumentos para una invocación de la función en particular.

```python
def saludo(leng):
    if leng == 'es':
        print('Hola')
    elif leng == 'fr':
        print('Bonjour')
    else:
        print('Boas tardes')
    
```

## Valores devueltos, de retorno
A menudo, una función tomará los argumentos, realizará algunos cálculos y devolverá un valor que se utilizará como valor de la llamada a la función, en la expresión de llamada, para tener un argumento de salida

La palabra clave **return** se usa para esto.

```python
def saludo():
    return "Hola"

print(saludo(), "Jon")
print(saludo(), "Kamela")
```

```python
def saludo(leng):
    if leng == 'es':
        return 'Hola'
    elif leng == 'fr':
        return 'Bonjour'
    else:
        return 'Boas tardes'
```

```python
print(saludo('en'),'Jon')
print(saludo('es'),'Kamela')
print(saludo('fr'),'Michael')

```

## Para evitar tener que depender del nombre de las variables internas:

```python
def escribe_media():
    media = (a + b) / 2
    print(f"La media de {a} y {b} es: {media}")
    return

a = 3
b = 5
escribe_media()
print("Programa terminado")
```

El problema de una función de este tipo es que es muy difícil de reutilizar en otros programas o incluso en el mismo programa, ya que sólo es capaz de hacer la media de las variables "a" y "b". Si en el programa no se utilizan esos nombres de variables, la función no funcionaría.

```python
def escribe_media(x, y):
    media = (x + y) / 2
    print(f"La media de {x} y {y} es: {media}")
    return

a = 35
b = 67
escribe_media(a, b)
print("Programa terminado")
```

```python
Para poder capturar, hacemos que la función resuelva la salida y la metemos en una varíable
```

```python
def calcula_media(x, y):
    resultado = (x + y) / 2 # la guardamos
    return resultado # lo ejecutamos

a = 890
b = 170
media = calcula_media(a, b)
print(f"La media de {a} y {b} es: {media}")
print("Programa terminado")
```

## Múltiples parámetros / argumentos

- Podemos definir más de un parámetro en la definición de la función.
- Simplemente agregamos más argumentos cuando llamamos a la función
- Hacemos coincidir el número y el orden de argumentos y parámetros.

```python
def suma(a, b):
    suma = a + b
    return suma

x = suma(3, 5)
print(x)
```

```python
def calcula_media(*argu):
    total = 0
    for i in argu:
        total += i
    resultado = total / len(argu)
    return resultado

a, b, c = 3, 5, 10
media = calcula_media(a, b, c)
print(f"La media de {a}, {b} y {c} es: {media}")
print("Programa terminado")
```
